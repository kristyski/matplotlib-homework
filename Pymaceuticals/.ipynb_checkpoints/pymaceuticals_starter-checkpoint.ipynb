## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sts

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merged_file = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
merged_file
```

```python
# Checking the number of mice.
mice_num = merged_file["Mouse ID"].nunique()
mice_num
```

```python
# All the data for the duplicate mouse ID.. 
finding_dups = merged_file[merged_file.duplicated(subset=["Mouse ID", "Timepoint"], keep=False)] #.count() gave me 10 rows
print(finding_dups)
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
renamed_merge_file_df = merged_file.rename(columns={"Mouse ID":"Mouse_ID"})
renamed_merge_file_df.head()

clean_df = renamed_merge_file_df.drop(renamed_merge_file_df[renamed_merge_file_df.Mouse_ID=="g989"].index)
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mice_num = clean_df["Mouse_ID"].nunique()
clean_mice_num
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method produces everything in a single groupby function

cleaner_df = clean_df.rename(columns={"Tumor Volume (mm3)":"Tumor_Volume"})
# cleaner_df.head()

agg_stats = cleaner_df.groupby("Drug Regimen").Tumor_Volume.agg(["mean", "median", "var", "std", "sem"])
print(agg_stats)
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pandas.
count_timepoints = cleaner_df.groupby("Drug Regimen").Timepoint.count().sort_values()
print(count_timepoints)
```

```python
panda_bar = count_timepoints.plot(kind="bar", x="Drug Regimen", y="count_timepoints", color="red", width=0.5, align="center",
                                  title="Number of Data Points per Drug Regimen")
plt.xlabel('Drug Regimen')
plt.ylabel('Data Points')
plt.show()
```

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pyplot.

matplotlib_df = pd.DataFrame({"Timepoints": count_timepoints})
matplotlib_df.sort_values("Timepoints", ascending=True).plot.bar(color='g',alpha=0.5, align="center")
plt.title("Number of Data Points per Drug Regimen")
plt.ylabel("Data Points")
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas
count_mice_sex = cleaner_df.Sex.value_counts()
panda_pie = count_mice_sex.plot(kind="pie", title="Female vs. Male Test Subjects", autopct='%1.1f%%')
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Determine outliers using upper and lower bounds
cap_quartiles = Capomulin["Tumor_Volume"].quantile([.25, .5, .75])
cap_lowerq = round(cap_quartiles[0.25],0)
cap_upperq = round(cap_quartiles[0.75],0)
cap_iqr = cap_upperq-cap_lowerq
print(f"Capomulin:")
print(f"Lower quartile: {cap_lowerq}")
print(f"Upper quartile: {cap_upperq}")
print(f"Inner quartile range: {cap_iqr}")

cap_lower_bound = cap_lowerq - (1.5*cap_iqr)
cap_upper_bound = cap_upperq + (1.5*cap_iqr)
print(f"Values below {cap_lower_bound} could be outliers.")
print(f"Values above {cap_upper_bound} could be outliers.")
print("--------------------------------")

ram_quartiles = Ramicane["Tumor_Volume"].quantile([.25, .5, .75])
ram_lowerq = round(ram_quartiles[0.25],0)
ram_upperq = round(ram_quartiles[0.75],0)
ram_iqr = ram_upperq-ram_lowerq
print(f"Ramicane:")
print(f"Lower quartile: {ram_lowerq}")
print(f"Upper quartile: {ram_upperq}")
print(f"Inner quartile range: {ram_iqr}")

ram_lower_bound = ram_lowerq - (1.5*ram_iqr)
ram_upper_bound = ram_upperq + (1.5*ram_iqr)
print(f"Values below {ram_lower_bound} could be outliers.")
print(f"Values above {ram_upper_bound} could be outliers.")
print("--------------------------------")

inf_quartiles = Infubinol["Tumor_Volume"].quantile([.25, .5, .75])
inf_lowerq = round(inf_quartiles[0.25],0)
inf_upperq = round(inf_quartiles[0.75],0)
inf_iqr = inf_upperq-inf_lowerq
print(f"Infubinol:")
print(f"Lower quartile: {inf_lowerq}")
print(f"Upper quartile: {inf_upperq}")
print(f"Inner quartile range: {inf_iqr}")

inf_lower_bound = inf_lowerq - (1.5*inf_iqr)
inf_upper_bound = inf_upperq + (1.5*inf_iqr)
print(f"Values below {inf_lower_bound} could be outliers.")
print(f"Values above {inf_upper_bound} could be outliers.")
print("--------------------------------")

cef_quartiles = Ceftamin["Tumor_Volume"].quantile([.25, .5, .75])
cef_lowerq = round(cef_quartiles[0.25],0)
cef_upperq = round(cef_quartiles[0.75],0)
cef_iqr = cef_upperq-cef_lowerq
print(f"Ceftamin:")
print(f"Lower quartile: {cef_lowerq}")
print(f"Upper quartile: {cef_upperq}")
print(f"Inner quartile range: {cef_iqr}")

cef_lower_bound = cef_lowerq - (1.5*cef_iqr)
cef_upper_bound = cef_upperq + (1.5*cef_iqr)
print(f"Values below {cef_lower_bound} could be outliers.")
print(f"Values above {cef_upper_bound} could be outliers.")
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
scatterdata = cleaner_df.loc[cleaner_df["Drug Regimen"]=="Capomulin"]
scatterdata2 = scatterdata.groupby(["Mouse_ID"]).mean()
scatterdata2

x_axis = scatterdata2["Weight (g)"]
y_axis = scatterdata2["Tumor_Volume"]

plt.scatter(x_axis, y_axis, marker="o", facecolors="red", edgecolors="grey", alpha=0.75)

plt.title("Drug Regimen: Capomulin")
plt.xlabel("Subject Weight (g)")
plt.ylabel("Tumor Volume (mm3)")

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
corr = round(sts.pearsonr(scatterdata2["Weight (g)"],scatterdata2["Tumor_Volume"])[0],2)
model = sts.linregress(scatterdata2["Weight (g)"],scatterdata2["Tumor_Volume"])

plt.scatter(scatterdata2["Weight (g)"],scatterdata2["Tumor_Volume"])
y_values = scatterdata2["Weight (g)"] * model[0] + model[1]

plt.plot(scatterdata2["Weight (g)"],y_values,color="red")
plt.title("Drug Regimen: Capomulin")
plt.xlabel("Subject Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
```
